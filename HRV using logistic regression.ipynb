```python
#Importing required libraries 
import pandas as pd
import numpy as np
%matplotlib notebook
import matplotlib.pyplot as plt
```

```python
#Reading the csv data and removoing the count column from data
train = pd.read_csv("train.csv").drop(columns = "datasetId")
test = pd.read_csv("test.csv").drop(columns = "datasetId")
```

```python
#Printing the test and train data 
train
```

```python
#Indexing the array for small dataset 
test1 = test[:20000]
train1 = train[:200000]
```

```python
#removing the tags of column in data from dataset

X_train = train1.drop(columns='condition')
Y_train = train1['condition']
X_test = test1.drop(columns='condition')
Y_test = test1['condition']

print(X_train.shape)
print(Y_test.shape)
```

```python
#spliting the data for training and testing 
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(X_train,Y_train,test_size=0.25,random_state=2)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
#Using scikit function logistic Regrssion and setting max itearator and fitting the data for training model
LG=LogisticRegression(max_iter=1000000)
LG.fit(x_train,y_train)
```

```python
print(LG.predict(X_test[0:10]))
```

```python
predictions=LG.predict(X_test)
```

```python
#Accuracy of trained model
score=LG.score(X_test,Y_test)
print(score*100)
```
